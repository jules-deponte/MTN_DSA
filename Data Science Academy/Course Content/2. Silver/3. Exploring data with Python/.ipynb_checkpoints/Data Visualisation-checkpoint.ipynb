# Visualising data in Python

Visualising data in Python is different from visualising data in Power BI. Power BI, generally, will be used to create reports for business decision makers. Visualising data in Python is useful for analysts to uncover insights in their data. 

There is a package available in Python for visualisation called Matplotlib. Practically all of the other visualisation packages for Python are built on top of Matplotlib. As other visualisation packages rely on Matplotlib, we will need to import this whenever we expect to carry out visualisations in our notebook. We import **matplotlib.pyplot as plt**. We will not spend any time working with Matplotlib itself, as there is another package, called Seaborn, which makes visualising data much easier. It is recommended that you go through the documentation of Matplotlib and Seaborn in your own time.

Matplotlib: https://matplotlib.org/

Seaborn: https://seaborn.pydata.org/

By convention, Seaborn is aliased as **sns**. In this notebook, we will be using some built-in datasets that come with the Seaborn package. As always, we will import Pandas as well.

The **%matplotlib inline** command is used to display the visual within the notebook iteself.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

There are several datasets that are preloaded in the Seaborn library which wwe can use to learn how to use the various plotting functions. To get a list of these datasets, run the **sns.get_dataset_names()** function.


```python
sns.get_dataset_names()
```

To load a dataset, we run the **sns.load_dataset()** function. We will load the **tips** dataset, and save it as a dataframe **df**.

```python
df = sns.load_dataset('tips')
```

Have a look at the first 10 rows of df.

```python
df.head(10)
```

## Visualising Numerical Data

This dataset contains data about customers at a restaurant, what their total bill was, the tip they left the server, the type of meal (lunch/dinner) and the number of people who ate there.

Something we may want to know is what the distribution of the total_bill is. To do this, we can use the built-in Pandas visualisation functions.

```python
df.hist('total_bill')
```

The above visual is built using the matplotlib package. There is some customisation you can do with this, but a far more user-friendly package is the Seaborn package.

To show the distribution of **total_bill**, we can call the **sns.distplot()** function. The displot visual only shows a single variable in the form of a histogram. This gives us a good indication of the distribution of the price of the bills, with the majority around $20. We can add an additional argument to change the size of the bins in the histogram. 

```python
sns.distplot(df['total_bill'], bins = 15)
```

If you wanted to compare two variables against onee another, for example, seeing how total_bill varies with respect to the size of the tip left for the server, we can use several different functions.

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = df)
```

```python
sns.jointplot(x='total_bill', y = 'tip', data = df)
```

```python
sns.pairplot(data = df, x_vars='total_bill', y_vars='tip', height = 5)
```

**Important**
When comparing 

If you want to compare all numeric columns against each other, you can use the pairplot function again, but this time, do not specify the x_vars and y_vars arguments.

You can further investigate any relationships by setting the **hue** argument to be one of the categories in the dataset. For example if we wanted to break the graphs down by time, add **time** and the **hue** argument.

```python
sns.pairplot(df, hue = 'time')
```

We can see there seems to be a relationship between **total_bill** and **tip**. We can call the **lmplot()** function to see what relationship exists. Also, we can add the **hue** argument again to compare different categories.

```python
sns.lmplot(x='total_bill', y = 'tip', data = df, hue = 'smoker',)
```

We can add a further variable to the plot by using the **col** argument in the **sns.lmplot()** function.

```python
sns.lmplot(x='total_bill', y = 'tip', data = df, hue = 'smoker', col='time')
```

We can show yet another category by using the **row** argument.

```python
sns.lmplot(x='total_bill', y = 'tip', data = df, hue = 'smoker', col='time', row = 'sex')
```

## Visualising Categorical Data

In the above plots, we were comparing numerical data against each other, with the optional addition of a single category. What if we wanted to compare several categories at once? There are some functions available in Seaborn for that.

Imagine we wanted to know how many patrons of each gender visited the restaurant by day. We can do a **sns.countplot()**. We can set the x- axis to **day** (which will show the days along the bottom of the graph), and the height of the bars will equal the count of the patrons for that day. If we set the **hue** argument to **sex**, for example, we can set how the count is made up of males and females.

```python
sns.countplot(x = 'day', data = df, hue='smoker')
```

The barplot looks similar to the countplot, but this is showing an agregation of data. For example, if we take the barplot of **tip**, we can see the average tip left per day. 

```python
sns.barplot(x = 'day', y = 'tip', data = df, hue='sex')
```

The **barplot()** function has an **estimator** argument, which, by default, is set to **mean**. If we set this to **sum**, we will see a different agregation applied to the visual.

```python
sns.barplot(x = 'day', y = 'tip', data = df, hue='sex', estimator=sum)
```

To visualise statistical information about the dataset, we can use the **sns.boxplot()** function. This creates a box and whisker plot, where the bounds of the box shows you the interquartile range, the line in the middle of the box shows the median, the whicskers show the min and max (1.5 * IQR), as well as the outliers. This is a very useful graph as it shows a lot of information in a condensed area.

```python
sns.boxplot(x = 'day', y = 'tip', data = df, hue='sex')
```

If we wanted to show more than two categories at the same time, we can use the **sns.PairGrid()**.
The PairGrid plot type allows for more flexibility than the pairplot, in that, you can specify the types of visuals you want for the upper part of the grid, the lower part and the diagonal, as shown in the cells below.

If you simply call **sns.PairGrid()**, you will simply end up with a grid of blank visuals.

```python
sns.PairGrid(data = df, hue='smoker')
```

If you want to populate the grid with data, you need to map the various sections of the grid to contain various visuals, as shown below.

```python
g = sns.PairGrid(data = df, hue='sex')
g.map_upper(sns.scatterplot)
g.map_diag(sns.kdeplot, shade = True)
g.map_lower(sns.scatterplot)
```

For small datasets, this PairGrid is fine, but when there are many numeric columns, it is impractical to visualise all of them in this manner. We can add some arguments to the function to limit the number of columns we show in the PairGrid.

In the example 

```python
g = sns.PairGrid(data = df.dropna(), 
                 x_vars = 'tip', 
                 y_vars= df.select_dtypes(include='category').columns, 
                 aspect=2.5)
g.map(sns.barplot)
```
