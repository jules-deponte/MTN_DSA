```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import time

from sklearn.linear_model import LinearRegression as lg

%matplotlib inline

df = sns.load_dataset('mpg')
df = df.drop(['origin', 'name'], axis = 1)
df[0] = 1
# df = df[[0, 'horsepower', 'displacement']]

df.dropna(inplace = True)
df.reset_index(inplace = True)
df.drop('index', axis = 1, inplace = True)

x = []
alpha = 0.00000001

y = df['displacement']
x = df.drop('displacement', axis = 1)

theta = []
temp = []

for i in range(0, len(x.columns)):
    theta.append(0)
    temp.append(999)
    
print(theta)
print(temp)
```

```python
x_mat = np.matrix(x)
theta_vec = np.matrix(theta).transpose()
y = np.matrix(y).transpose()
```

```python
x_mat
```

```python
theta_vec
```

```python
for i in range(0, 1000):
    error = ((x_mat * theta_vec) - y)
    error_product = np.sum(np.multiply(error, np.matrix(x)),axis = 0)
    error_product = alpha * error_product / len(df)
    theta_vec = (theta_vec.transpose() - error_product).transpose()
    i = i + 1


print(error)
```

```python
df['error'] = error
```

```python
theta_vec
```

```python
df['pred'] = x_mat * theta_vec
```

```python
model = lg()
model.fit(x, y)
predictions = model.predict(x)
model.coef_
```

```python
df['pred_skl'] = predictions
```

```python
plt.figure(figsize=(8,8))
plt.scatter('horsepower', 'displacement', marker = 'x', data = df)
plt.plot('horsepower', 'pred', data = df)
# plt.plot('horsepower', 'pred_skl', data = df)
plt.legend()
plt.xlabel('horsepower')
plt.ylabel('displacement')
```
