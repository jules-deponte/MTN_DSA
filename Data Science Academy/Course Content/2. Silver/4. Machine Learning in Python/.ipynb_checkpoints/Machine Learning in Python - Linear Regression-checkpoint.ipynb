#### Galton's data on the heights of parents and their children, by child
#### Description
This data set lists the individual observations for 934 children in 205 families on which Galton (1886) based his cross-tabulation.

In addition to the question of the relation between heights of parents and their offspring, for which this data is mainly famous, Galton had another purpose which the data in this form allows to address: Does marriage selection indicate a relationship between the heights of husbands and wives, a topic he called assortative mating?

Usage
data(GaltonFamilies)
Format
A data frame with 934 observations on the following 8 variables.

**family**
family ID, a factor with levels 001-204

**father**
height of father in inches

**mother**
height of mother

**midparentHeight**
mid-parent height, calculated as (father + 1.08 * mother)/2 in inches

**children**
number of children in this family

**childNum**
number of this child within family. Children are listed in decreasing order of height for boys followed by girls

**gender**
child gender, a factor with levels female male

**childHeight**
height of child in inches

## Objective

We would like to create a machine learning model that predicts teh height of the child, given the height of the parents.

Let us visualise the data. We are interested in the relationship between the height of the children and the height of the parents. We can therefore drop the other columns from the pairplot.

```python
sns.pairplot(df.drop(['Case', 'children', 'childNum'], axis = 1), hue='gender')
```

We would like to make sure that the data types are all correct.

```python
df.dtypes
```

The gender is an **object** type. This is a categorical variable, and we would like to change it as such. Recall from the previous lesson how to do this.

```python
# First change the data type to 'category'.
df['gender'] = df['gender'].astype('category')

# Then create a dictionary which shows the category code and the category label. Print this dictionary
gender_cat = dict(enumerate(df['gender'].cat.categories))
print(gender_cat)

# Reassign the gender column to the category codes, such that the machine learning algorithm can work with it.
df['gender'] = df['gender'].cat.codes
```

## Machine Learning

We will now import the machine learning model from Scikit Learn. Scikit Learn is a massive library, and we do not want to load it in its entirity into our workspace. Therefore, we will only import the relevant models from scikit learn. 

As linear regression is a linear model, we will use the **linear_model** sub-library. We run the command 
**from sklearn.linear_model import LinearRegression**. We can aliase it as **lg**.

We can then create an object called **model**, which we set equal to the Linear Regression algorithm. We will fit our data to this model, and use it to run predictions.

```python
from sklearn.linear_model import LinearRegression as lg
model = lg()
```

In order to run the machine learning, we need to create a set of data with the attributes and another set with the labels. By convention, the attributes we call 'x' and the labels we call 'y'.

In our case, we would like to try predict the height of the child, given the height of their parents.

For the attributes set (x), we are only interested in the numerical columns. We are not interetsed in the ID columns, such as **family** and **Case**. The number of children and the number of the child in the family does not matter, so we remove these columns from our set.

For the labels set (y), we are only interested in the childHeight.

```python
x = df.drop(['family', 'Case', 'children', 'childNum', 'childHeight'], axis = 1)
y = df['childHeight']
```

With our datasets ready, we can now fit the data to our model. This is the actual 'machine learning'. The **model** is running the Linear Regression algorithm in the background. After a short time, it will have created a model against which we can run predictions.

```python
model.fit(x, y)
```

We can now run a prediction against our model. to do this, we can call the **.predict()** function on the **model** object. The output is an array of heights that were predicted by the linear regression model.

```python
pred = model.predict(x)
pred
```

We can save this array into the dataframe to get a first-pass feel for how close the predictions are to the actual values. We can create another column which calculates the error between each prediction and each actual value.

```python
df['pred'] = pred
df['error'] = df['pred'] - df['childHeight']
df.head(10)
```

The values look quite good. 

We can try to visualise the predictions by plotting them on a scatter plot. In the plot below, we have the actual values given as 'x's and the predictions given as triangles. We can see that there is a fair amount of overlap between the predictions and actual values.

```python
plt.figure(figsize=(8, 8))
colors = {0:'red', 1:'blue'}
plt.scatter(df['father'], df['childHeight'], marker = 'x', c = df['gender'].apply(lambda x: colors[x]))

colors = {0:'pink', 1:'black'}
plt.scatter(df['father'], df['pred'], marker = '^', c = df['gender'].apply(lambda x: colors[x]))

plt.xlabel('Father')
plt.ylabel('Child')
```

We want to quantify the accuracy of the model. To do this, we will use another Scikit Learn library called **metrics**. This contains formulas that can quickly give us numerical information as to the accuracy of our model.

```python
from sklearn import metrics
```

We can visualise the Mean Absolute Error, and the Root of the Mean Squared Error. Both of these give an indication of how 'wrong' an average prediction would be.

```python
print("Mean Absolute Error:", str(metrics.mean_absolute_error(y, pred)))
print("Root Mean Squared Error:", str(np.sqrt(metrics.mean_squared_error(y, pred))))
print('\n')
print("Mean Error:", str(np.mean(df['error'])))
print("Error standard deviation:", str(np.std(df['error'])))
```

We can see that the errors are in the range of 1.7 to 2.2 inches. This means that, for a predicition of our machine learning model, we can predicit the height of the child, given the height of the parents, within these bounds.

We can plot a histogram of the errors. We can see that they are normally distributed and are centered around zero. From the metrics above, we can see that the mean error is 0.27" and the standard deviation is 2.18". This means that the average prediction error made by the model has a 66.67% chance of being 0.27", plus or minus 2.18".

```python
sns.distplot(df['error'], bins = 25, kde = False)
```

## Using a test set

All of the predictions we made in the previous section were made using the training dataset. In order to measure the performance of the model, we can use a test set. We will process the data in exactly the same way we did previously; we will use the same colums, and we will change the gender column to categorical as well.

```python
df_test = pd.read_csv(r"C:\Users\julesd\OneDrive - Altron Karabina\Karabina\Demos and PoC's\MTN\Data Science Academy\Course Content\2. Siver\Datasets\GaltonFamiliesTest.csv")
df_test
```

```python
# First change the data type to 'category'.
df_test['gender'] = df_test['gender'].astype('category')

# Then create a dictionary which shows the category code and the category label. Print this dictionary
gender_cat = dict(enumerate(df_test['gender'].cat.categories))
print(gender_cat)

# Reassign the gender column to the category codes, such that the machine learning algorithm can work with it.
df_test['gender'] = df_test['gender'].cat.codes
```

```python
x_test = df_test.drop(['family', 'Case', 'children', 'childNum', 'childHeight'], axis = 1)
y_test = df_test['childHeight']
```

This is where we use the model which we have previously defined, and make predictions on a dataset which the machine learning model has not seen before.

```python
pred_test = model.predict(x_test)
```

We can now write this pred_test column to our new df_test dataframe. We can also work out the errors.

```python
df_test['pred_test'] = pred_test
df_test['error'] = df_test['pred_test'] - df_test['childHeight']
df_test
```

We can visualise the data, once again.

```python
plt.figure(figsize=(8, 8))
colors = {0:'red', 1:'blue'}
plt.scatter(df_test['father'], df_test['childHeight'], marker = 'x', c = df_test['gender'].apply(lambda x: colors[x]))

colors = {0:'pink', 1:'black'}
plt.scatter(df_test['father'], df_test['pred_test'], marker = '^', c = df_test['gender'].apply(lambda x: colors[x]))
```

And we can get a sense of the errors produced.

```python
print("Mean Absolute Error:", str(metrics.mean_absolute_error(y_test, pred_test)))
print("Mean Squared Error:", str(metrics.mean_squared_error(y_test, pred_test)))
print('\n')
print("Mean Error:", str(np.mean(df_test['error'])))
print("Error standard deviation:", str(np.std(df_test['error'])))
```

We can see that the performance metrics of the predictions made on the test set are similar to those made on the training set. 

```python
sns.distplot(df_test['error'], bins = 25, kde = False)
```
