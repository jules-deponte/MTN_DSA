```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python

df = pd.DataFrame(columns = ['x', 'x2', 'x3'])
```

```python

df = pd.DataFrame(columns = ['x', 'x2', 'x3'])
x = []
x2 = []
x3 = []
x4 = []
x5 = []
x6 = []
x0 = []


for i in range(-5, 8):
    x.append(i)
    x2.append(i**2 - 2*i + 3)
    x3.append(2*i - 1)
    x4.append(2)
    x5.append(6*i - 13)
    x6.append(-10*i -13)
    x0.append(0)
    
df['x'] = x
df['x2'] = x2
df['x3'] = x3
df['x4'] = x4
df['x5'] = x5
df['x6'] = x6
df['x0'] = x0
```

```python
df
```

```python
plt.figure(figsize=(8,8))
plt.plot(x, x2)
# plt.plot(x, x3)
plt.plot(x, x4) #Minimum
plt.plot(x, x5, c = 'g')
# plt.plot(x, x6, c = 'r')
plt.plot(x, x0, color = 'black')
plt.plot(x0, x5, color = 'black')
plt.xlabel('Coefficients [a]')
plt.ylabel('error [J]')
plt.xlim((-6, 8))
plt.ylim((-10,40))
plt.legend()
```

```python
df = pd.DataFrame()

x = []
x_count = 0
hx1 = []
hx2 = []
x0 = []

for i in range(0,100):
    hx1.append(-np.log(x_count))
    hx2.append(1-(np.log(1-x_count)))
    x.append(x_count)
    x0.append(0)
    x_count = x_count + 0.01
    
df['x'] = x
df['hx2'] = hx2
df['hx1'] = hx1
```
